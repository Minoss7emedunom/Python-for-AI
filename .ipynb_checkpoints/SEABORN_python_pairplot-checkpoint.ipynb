## SEABORN function

Almoost all Seaborn functions have the same structure

$sns.function(x,y, data , hue, size ,style)$

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.drop(['alone', 'alive', 'who', 'adult_male', 'embark_town', 'class'], axis = 1, inplace = True)
titanic.dropna(axis = 0, inplace = True)
titanic.head()
```

```python
sns.pairplot(titanic)
```

```python
sns.catplot(x= 'pclass', y = 'age', data = titanic, hue = 'sex')
```

```python
sns.boxplot(x = 'pclass', y = 'age', data = titanic, hue = 'sex')
```

```python
sns.distplot(titanic['fare'])
```

```python
sns.jointplot(x = 'age',y = 'fare', data = titanic, kind = 'hex')
```

```python
titanic['sex'] = titanic['sex'].astype('category').cat.codes
titanic['embarked'] = titanic['embarked'].astype('category').cat.codes
titanic['deck'] = titanic['deck'].astype('category').cat.codes
```

```python
sns.heatmap(titanic.corr())
```
