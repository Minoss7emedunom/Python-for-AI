## Sklearn Supervised Learning

```python
import numpy as np
import matplotlib.pyplot as plt
```

## (1) Regression avec Sklearn


```python
np.random.seed(0)
m = 100 # creating 100 samples
X = np.linspace(0,10,m).reshape(m,1)
y = X + np.random.randn(m,1)
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,y)
model.score(X,y)

prediction = model.predict(X)

plt.scatter(X,y)
plt.plot(X, prediction , c='r')
```

```python
X = np.linspace(0,10,m).reshape(m,1)
y = X**2 + np.random.randn(m,1)
```

```python
plt.scatter(X,y)
```

```python
model = LinearRegression()
model.fit(X,y)
model.score(X,y)

prediction = model.predict(X)

plt.scatter(X,y)
plt.plot(X, prediction , c='r')
```

```python
from sklearn.svm import SVR 
```

```python
model = SVR(C = 100)
model.fit(X,y)
model.score(X,y)

prediction = model.predict(X)

plt.scatter(X,y)
plt.plot(X, prediction , c='r' , lw =3)
```

## (2) Sklearn Classification KNN

```python
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
print(titanic.shape)
titanic.head()
```

```python
titanic = titanic[['survived', 'pclass', 'sex','age']]
titanic.dropna(axis =0 , inplace= True)
titanic['sex'].replace(['male','female'], [0,1], inplace = True)
titanic.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

model = KNeighborsClassifier()

```python
y = titanic['survived']
X = titanic.drop('survived', axis = 1)
```

```python
print(y)
```

```python
print(X)
```

```python
model.fit(X,y)
model.score(X,y)
```

```python
model.predict(X)
```

```python
def survive(model , pclass = 3 , sex = 0 , age = 26):
    x = np.array([pclass,sex,age]).reshape(1,3)
    print(model.predict(x))
    print(model.predict_proba(x))
```

```python
survive(model)
```

# Exercise
write algorithm to find the best  n_neighbors

```python
score = []
best_k = 1
best_score = 0

for k in range(best_k , 30):
    model = KNeighborsClassifier(n_neighbors = k)
    model.fit(X,y)
    score.append(model.score(X,y))
    
    if best_score < model.score(X,y):
        best_k = k
        best_score = model.score(X,y)
        
print(best_k)
plt.plot(score)
```
