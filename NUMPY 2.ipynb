## SUB SETTING

```python
B = A[0:2,0:2]
print(B)
```

```python
A[0:2,0:2] = 10
print(A)
```

```python
print(A[:,1:])
```

```python
print(A[:,-2:])
```

```python
B = np.zeros((4,4))
print(B)
```

```python
B[1:3,1:3] = 1
C = B[1:3,1:3]  
print(C)
```

```python
print(B)
```

```python
D = np.zeros((5,5))
print(D)
```

## BOOLEAN INDEXING

```python
#np.random.seed(0)
A= np.random.randint(0,10, [5,5])
print(A)
```

```python
print(A<5)
```

```python
A[A<5] = 10
print(A)
```

```python
A[(A<5) & (A>2)] = 10
print(A)
```

```python
B = np.random.randint(0,255, [1024,720])
print(B)
```

```python
B[B>200] = 255
print(B)
```

```python
A= np.random.randint(0,10, [5,5])
print(A[A<5])
#print(A)
```

```python
print(A[A<5].shape)
```

```python
B = np.random.randn(5,5)
print(B)
print(B[A<5])
```

```python
print(B[A<5].shape)
```

## EXERCICE

```python
from scipy import misc
import matplotlib.pyplot as plt
face = misc.face(gray = True)
plt.imshow(face, cmap = plt.cm.gray)
plt.show()
```

```python
print(type(face))
```

```python
print(face.shape)
```

```python
larg= face.shape[0]
long = face.shape[1]
face1 = face
face1 =face1[(larg//4): -larg//4 , long//4 : -long//4]
face1[face1>200] = 255
plt.imshow(face1, cmap = plt.cm.gray)
plt.show()
```

```python
face = face[::2,::2]
plt.imshow(face, cmap = plt.cm.gray)
plt.show()
```
