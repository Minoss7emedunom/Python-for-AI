## Matplotlib Fonction plot

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,2,10)
y = x**2
print(x)
```

```python
plt.plot(x,y)
plt.show()
```

## X et Y doivent avoir les même dimensions

```python
plt.scatter(x,y)
plt.show()
```

## Matplotlib styles

```python
plt.plot(x,y, c = 'black', lw = 5, ls = '--')
plt.show()
```

## Matplotlib cycle de vie

```python
plt.figure()
plt.plot(x,y , label = 'quadratique')
plt.plot(x,x**3, label = 'cubique')
plt.title('figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.show()
plt.savefig('figure.png')

plt.figure()
plt.plot(x,y)
```

## Matplotlib subplot

```python
plt.subplot(2,1,1)
plt.plot(x,y, c = 'red')
plt.subplot(2,1,2)
plt.plot(x,y, c ='blue')
```

```python
plt.figure()
plt.plot(x,y)
plt.title('graphique 1')
plt.subplot(2,2,2)
plt.plot(x , np.sin(x) , label = 'sinus')
plt.plot(x, np.cos(x) , label = 'cosinus')
plt.legend()
plt.show()
```

## Matplotlib OOP

```python
fig , ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
fig, ax = plt.subplots(2,1, sharex = True)
ax[0].plot(x,y)
ax[1].plot(x, np.sin(x))
plt.show()
```

## EXERCICE
Vous avez un dictionnaire qui comprend N datasets.
Créez une fonction qui trace N graphiques sur unne seule figure

```python
dataset = {f"experience{i}" : np.random.randn(100,3) for i in range(6)}
```

```python
def graphique(data):
    n = len(data)
    plt.figure(figsize = (20,13))
    for k,i in zip(data.keys(), range(1,n+1)):
        plt.subplot(n,1,i)
        plt.plot(data[k])
        plt.title(k)
    plt.show()
```

```python
graphique(dataset)
```
