## Graphiques les plus utiles avec matplotlib pour faire du marchine learning

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

x = iris.data
y = iris.target
names = list(iris.target_names)


print(f'x contient {x.shape[0]} exemples et {x.shape[1]} variables')
print(f'il y a {np.unique(y).size} classes')
```

```python
print(x)
```

```python
print(y)
```

## (5) Graphique de Classification avec plt.scatter

```python
plt.scatter(x[:,0], x[:,1], c = y , alpha = 0.5, s = x[:,2]*100)
plt.xlabel('Longueur sépal')
plt.ylabel('Largeur sépal')
plt.show()
```

# (4) Graphique 3D

```python

from mpl_toolkits.mplot3d import Axes3D
```

```python
ax = plt.axes(projection = '3d')
ax.scatter(x[:,0], x[:,1], x[:,2], c = y)
```

```python
f = lambda x,y : np.sin(x) + np.cos(x+y)
```

```python
X = np.linspace(0,5,100)
Y = np.linspace(0,5,100)
X,Y = np.meshgrid(X,Y)

Z = f(X,Y)
```

```python
print(Z.shape)
```

```python
ax = plt.axes(projection = '3d')
ax.plot_surface(X,Y,Z, cmap = 'plasma')
plt.show()
```

## (3) Histogramme

```python
plt.hist(x[:,0], bins = 20)
plt.hist(x[:,1], bins = 20)
plt.show()
```

```python
plt.hist2d(x[:,0], x[:,1], cmap = 'Blues')
plt.xlabel('Longueur sépal')
plt.ylabel('Largeur sépal')
plt.colorbar()
plt.show()
```

```python
from scipy import misc
face = misc.face(gray = True)
plt.imshow(face, cmap = 'gray')
```

```python
plt.hist(face.ravel(), bins = 255)
plt.show()
```

## (2) Contour plots

```python
plt.contour(X,Y,Z, 20, colors = 'black')
```

```python
f = lambda x,y : np.sin(x) + np.cos(x+y)*np.cos(x)

X = np.linspace(0,5,100)
Y = np.linspace(0,5,100)
X,Y = np.meshgrid(X,Y)

Z = f(X,Y)
```

```python
plt.contour(X,Y,Z, 20, colors= 'black')
```

```python
plt.contourf(X,Y,Z, 20, cmap = 'RdGy')
plt.colorbar()
```
